```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
```

```python


engine = create_engine("sqlite:///restaurants.sqlite")
```

```python
# Create a connection to the engine called `conn`

conn = engine.connect()
```

```python
# Use `declarative_base` from SQLAlchemy to model the demographics table as an ORM class
# Make sure to specify types for each column

Base = declarative_base()

class RestaurantCount(Base):
    __tablename__ = 'restaurant_count'

    zipcode = Column(Text, primary_key=True)
    Restaurant_Count_2012 = Column(Integer)
    Restaurant_Count_2013 = Column(Integer)
    Restaurant_Count_2014 = Column(Integer)
    Restaurant_Count_2015 = Column(Integer)
    Restaurant_Count_2016 = Column(Integer)

    def __repr__(self):
        return f"id={self.id}, name={self.name}"
```

```python
# Use `create_all` to create the demographics table in the database
Base.metadata.create_all(engine)
```

```python
# Use MetaData from SQLAlchemy to reflect the tables
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
table = sqlalchemy.Table('restaurant_count', metadata, autoload=True)
```

```python
# Use `table.delete()` to remove any existing data.
conn.execute(table.delete())
```

```python
df.to_sql("restaurant_count", con=engine, if_exists='replace')
```
